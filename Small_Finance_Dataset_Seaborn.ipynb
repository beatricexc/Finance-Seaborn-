```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/Finance_data.csv")
```

```python
#correlation between savings objective and gender
sns.set_theme(style="darkgrid")
sns.countplot(x="What are your savings objectives?",hue="gender",data= data,palette="Set3" )
```

```python
#Boxplot of correlation btween age and equity market
sns.boxplot(x="age", y="Equity_Market", data=data, palette = "gist_ncar")
```

```python
#Correlation of gender between other parameters
sns.pairplot(data, hue="gender", palette ="Set3")
```

```python
sns.jointplot(x="age", y="Equity_Market", data= data, kind="kde", color="blue", height=7)
```
